```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
```

```python
# fetching data from twitter
df_city = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    'global warming near:"Berlin" within:1000km').get_items(),100000))
```
